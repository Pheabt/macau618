```python
import requests
import json
import pprint
import pandas as pd
import pprint
```

```python
key='d3b9d9e62888e1c449294703009cc1b2' 
def getlocation(address):
    url='https://restapi.amap.com/v3/geocode/geo'
    params = {
        'key':key,
        'address':address
    }
    res = requests.get(url, params)
    json_data = json.loads(res.text)
    return json_data
```
